```python
query = """
        select a.Arrest_Num, a.Arrest_Datetime, a.Age_TOA, a.Gender, a.Race, a.Ethnicity, a.ArrestLocationst_NameCityStateZip, 
        a.Impaired, a.Weapon_Poss, a.VisualProbationViolation, a.Warrant, a.GovernorsWarrant, 
        b.txt_offense_code, b.txt_offense_type, b.txt_offense_short_description, b.txt_probable_cause,
        b.txt_warrant_type
        from arr_det a
        join pm_rev b on a.Arrest_Num = b.Arrest_Num"""
total_data = sqldf(query, locals())
total_data.head()
```

```python
print(total_data.shape)
```

```python
query = """
        select a.zip_cd, a.Arrest_Num, b.Pop
        from zip_arr_cnts a
        join demo b on a.zip_cd = b.Zip
        group by a.zip_cd"""
arr_agg = sqldf(query, locals())
arr_agg.head()
```

```python
arr_agg['arrests_per'] = arr_agg['Arrest_Num']/arr_agg['Pop']
arr_agg.head()
```

```python
map_data3 = arr_agg.drop(['Arrest_Num', 'Pop'], axis = 1)
geo = r'C:\\Users\\grayson.b.absher\\Documents\\clt_zip.geojson'
map = folium.Map(location=[35.2271, -80.8431])
map.choropleth(geo_data = geo, data = map_data3,
               columns = ['zip_cd', 'arrests_per'], fill_color = 'YlOrRd', key_on = 'feature.properties.zipcode')
map
```
